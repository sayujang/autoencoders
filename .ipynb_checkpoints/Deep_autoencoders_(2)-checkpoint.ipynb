```python
class Encoder(nn.Module):
    def __init__(self,in_channels,latent_dim):
        super().__init__()
        self.in_channels=in_channels
        self.latent_dim=latent_dim
        self.encode=nn.Sequential(
            nn.Conv2d(in_channels,20*in_channels,kernel_size=4,stride=2,padding=1),#(3,32,32)->(6,16,16)#pytorch expects (b,c,h,w)  ie nchw format in all nn.layer format
            nn.ReLU(),
            nn.Conv2d(20*in_channels,42*in_channels,kernel_size=4,stride=2,padding=1),#(6,16,16)->(12,8,8)
            nn.ReLU(),
            nn.Conv2d(42*in_channels,85*in_channels,kernel_size=4,stride=2,padding=1),#(12,8,8)->(24,4,4)
            nn.ReLU(),
            nn.Flatten(), #will output : (batch, flattened_dimension)
            nn.Linear(4*4*85*in_channels,latent_dim)

        )
    def forward(self,x):
        encoding=self.encode(x)
        return encoding




```

```python
class Decoder(nn.Module):
    def __init__(self,in_channels,latent_dim):
        super().__init__()
        self.in_channels=in_channels
        self.latent_dim=latent_dim
        self.decode=nn.Sequential(
            nn.Linear(latent_dim,4*4*85*in_channels),
            nn.ReLU(),
            nn.Unflatten(dim=1, unflattened_size=(85*in_channels,4,4)),
            nn.ConvTranspose2d(85*in_channels,42*in_channels,4,stride=2,padding=1), #(24,4,4)->(12,8,8)
            nn.ReLU(),
            nn.ConvTranspose2d(42*in_channels,20*in_channels,4,stride=2,padding=1),#(12,8,8)->(6,16,16)
            nn.ReLU(),
            nn.ConvTranspose2d(20*in_channels,in_channels,4,stride=2,padding=1),#(6,16,16)->(3,32,32)
            nn.Sigmoid()
        )
    def forward(self,x):
        decoding=self.decode(x)
        return decoding
```

```python
class Autoencoder(nn.Module):
    def __init__(self,in_channels,latent_dim):
        super().__init__()
        self.in_channels=in_channels
        self.latent_dim=latent_dim
        self.encoder=Encoder(self.in_channels,self.latent_dim)# we are creating an object of Encoder class
        self.decoder=Decoder(self.in_channels,self.latent_dim)
    def forward(self,x):
        latent_output=self.encoder(x) #here we are calling __call__ method of Encoder object which under the hood runs forward() function
        recons_output=self.decoder(latent_output)
        return recons_output
```

```python
in_channels=3
latent_dim=384
model=Autoencoder(in_channels,latent_dim).to(device) #we instantiate pytorch model in training mode by default so model.train() is redundant
criterion=nn.MSELoss() #observation: MSEloss has much lower starting loss compared to L1loss but its generally due to scalling effect as square of initial losses(if loss<1) is always greater than the absolute ie MAE(L1loss)
optimizer=torch.optim.Adam(model.parameters(),lr=1e-3,weight_decay=1e-5)
```

```python
num_epochs=60
outputs=[]
for epoch in range(num_epochs):
    running_loss=0
    for (img,_) in train_loader:
        img=img.to(device, non_blocking=True)#pin_memory=True prepares the cpu tensors for fast gpu transfer while non_blocking=True transfers tensor to gpu asynchronously(cpu loading tensors and gpu transfer occur simultaneously)
        x_recons=model(img)
        loss=criterion(x_recons,img)
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        running_loss+=loss.item()*img.size(0) #img.shape/img.size=(64,3,32,32)# here loss.item() by default gives the average loss of images in a batch
    epoch_loss=running_loss/len(train_loader.dataset)
    print(f'Epoch:{epoch+1}, Loss:{epoch_loss:.5f}')
    outputs.append((epoch, img.cpu(), x_recons.detach().cpu()))
```

```python
#plotting images shere
epoch, orig_imgs, recon_imgs = outputs[-1]
n = 10 #no of images
plt.figure(figsize=(12, 4))
for i in range(n):
    ax = plt.subplot(2, n, i + 1) #original
    img = orig_imgs[i]
    # unnormalize from [-1,1] to [0,1] in case of tanh() as last layer
    #img = (img * 0.5) + 0.5 #inversefunction of normalization
    npimg = img.numpy().transpose((1, 2, 0))#matplotlib expects (h,w,c) and numpy arrays
    plt.imshow(np.clip(npimg, 0, 1)) #np.clip()<=(input_array,amin,amax) to avoid any round off error by denormalization
    ax.axis('off')
    if i == 0:
        ax.set_title('Original')

    # reconstructed
    ax = plt.subplot(2, n, n + i + 1)
    recon = recon_imgs[i]
    #recon = (recon * 0.5) + 0.5
    nprecon = recon.numpy().transpose((1, 2, 0))
    plt.imshow(np.clip(nprecon, 0, 1))
    ax.axis('off')
    if i == 0:
        ax.set_title('Reconstruction')

plt.suptitle(f'Epoch {epoch} Reconstructions', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
#random image recontruction:
"""
Model fails to generalize to random noise as it learned the structure of cifar10 dataset in its latent space. so reconstructs garbage image
"""
random=torch.rand((1,3,32,32)).to(device) #pytorch expects (b,c,h,w)  i.e. nchw format in all nn.layer format
with torch.no_grad():
  random_recons=model(random)

random=random.squeeze(dim=0) # torch.squeeze(dim), dim=(0,1,2,3) squeeze is possible only if at given dimensional value is 1 is can't squeeze (23,3,32,32) in this case
random_recons=random_recons.squeeze(dim=0)
plt.subplot(1,2,1)
plt.axis('off')
plt.imshow(random.cpu().numpy().transpose((1,2,0)))
plt.title("random noise")
plt.subplot(1,2,2)
plt.axis('off')
plt.imshow(random_recons.cpu().numpy().transpose((1,2,0)))
plt.title("reconstructed random image")
```

```python
#Finding visually similar images:
from tqdm import tqdm #it wraps around iterables to show progress bars ie percentage complete, estimated time to complete, etc
def embed_images(model,data_loader,device):
  Encoder_eval=model.encoder
  Encoder_eval.eval()
  img_list=[]
  embed_list=[]
  for img,_ in tqdm(data_loader,desc="Encoding..",leave=False):
    img=img.to(device) #(64,3,32,32)
    with torch.no_grad():
      z=Encoder_eval(img) #(64,latent_dim)
    img_list.append(img.cpu())
    embed_list.append(z.cpu())
  img_tot=torch.cat(img_list,dim=0)#torch.cat([tensor.size=(64,3,32,32),tensor.size=(64,3,32,32),....])==>(N,3,32,32) where N is total number of images in the train/test dataset
  embed_tot=torch.cat(embed_list,dim=0)#(N,latent_dim)
  return img_tot,embed_tot
```

```python
test_data=datasets.CIFAR10(root='./deepdata2',train=False, download=True, transform=transform)
```

```python
test_loader=DataLoader(dataset=test_data, batch_size=64, shuffle=True,
    num_workers=2,
    pin_memory=True)
```

```python
train_img, train_z=embed_images(model,train_loader,device)
test_img, test_z=embed_images(model,test_loader,device)
```

```python
def plot_similar_images(query_img,query_z,train_img,train_z,k=9):
  eucli_dist=torch.norm(train_z-query_z.unsqueeze(0), dim=1) #norm collapses dimension to [N]
  index=torch.argsort(eucli_dist)[:k]
  imgtoplot=torch.cat([query_img.unsqueeze(0),train_img[index]],axis=0)
  plt.figure(figsize=(12,3))
  for i in range(k+1):
    plt.subplot(1,k+1,i+1)
    plt.axis('off')
    plt.imshow(imgtoplot[i].numpy().transpose((1,2,0)))
  plt.show()
```
